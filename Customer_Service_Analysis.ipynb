```python
import pandas as pd 

# Loading a CSV file into Python so you can work with it
df = pd.read_csv('customer_shopping_behavior.csv')

# df is short for DataFrame
# df is a variable that holds the entire spreadsheet in memory
```

```python
# to check the summary statistics of all the columns (numerical and categorical)
df.describe(include='all') 

# if we do df.describe() -> get the summary statistics of numerical columns
```

```python
# check the missing values
df.isnull().sum()

# Review Rating has 37 null values --> we could replace it with mean or median 
# mean can get affected by outliers and skew the distribution of our data so, generally chose median
# median is robust to outliers

# NOT THE BEST PRACTICE as rating can differ based on the category
# (fill all the missing Review Rating with the overall median of the entire column)
# eg -> rating of clothing item and footwear item

# Input the missing values with median Review Rating within each category 
# eg -> if rating of clothing is missing then it is replaced with clothing's rating not footwear item
```

```python
# Input the missing values with median Review Rating within each category 
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

# "Cleaning it the right way!!"
```

```python
# check the missing values
df.isnull().sum()

# no NULL value remains
```

```python
# changing the column names to snakecase to remember it better
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

df = df
```
